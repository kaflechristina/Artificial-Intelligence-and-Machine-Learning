```python
# Task 1 : Create a Python program that converts between different units of measurement.
def convert_units(conversion_type, value):
    """
    Convert units based on the specified conversion type.

    Parameters:
    conversion_type (str): The type of conversion ('m_to_ft', 'ft_to_m', 'kg_to_lbs', 'lbs_to_kg', 'L_to_gal', 'gal_to_L')
    value (float): The numeric value to convert

    Returns:
    float: The converted value
    """
    try:
        if conversion_type == "m_to_ft":
            return value * 3.28084
        elif conversion_type == "ft_to_m":
            return value / 3.28084
        elif conversion_type == "kg_to_lbs":
            return value * 2.20462
        elif conversion_type == "lbs_to_kg":
            return value / 2.20462
        elif conversion_type == "L_to_gal":
            return value * 0.264172
        elif conversion_type == "gal_to_L":
            return value / 0.264172
        else:
            return "Unsupported conversion type"
    except Exception as e:
        return f"Error: {e}"

# Map options to conversion types
conversion_map = {
    1: "m_to_ft",
    2: "ft_to_m",
    3: "kg_to_lbs",
    4: "lbs_to_kg",
    5: "L_to_gal",
    6: "gal_to_L"
}

print("Choose a conversion type:")
print("1. m to ft")
print("2. ft to m")
print("3. kg to lbs")
print("4. lbs to kg")
print("5. L to gal")
print("6. gal to L")

# Keep asking until valid input is provided
while True:
    try:
        choice = int(input("Enter option number (1-6): "))
        if choice not in conversion_map:
            print("Invalid option! Please choose a number between 1 and 6.")
            continue  # ask again

        value = float(input("Enter the value to convert: "))
        result = convert_units(conversion_map[choice], value)
        print(f"Converted value: {result}")
        break  # exit loop after successful conversion
    except ValueError:
        print("Invalid input! Please enter numeric values.")
```

```python
# Task 2: Create a Python program that performs various mathematical operations on a list of numbers.

# Define functions for each operation

def calculate_sum(numbers):
    """Return the sum of a list of numbers."""
    return sum(numbers)

def calculate_average(numbers):
    """Return the average of a list of numbers."""
    return sum(numbers) / len(numbers)

def calculate_max(numbers):
    """Return the maximum value from a list of numbers."""
    return max(numbers)

def calculate_min(numbers):
    """Return the minimum value from a list of numbers."""
    return min(numbers)

# Prompt user to choose an operation
print("Choose an operation:")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

while True:
    try:
        choice = int(input("Enter option number (1-4): "))
        if choice not in [1, 2, 3, 4]:
            print("Invalid option! Please choose 1, 2, 3, or 4.")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter a number.")

# Ask user how many numbers they want to process (must be >=2)
while True:
    try:
        n = int(input("How many numbers do you want to process? "))
        if n < 2:
            print("Please enter a number greater than 1.")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter a valid number.")

# Collect the numbers one by one
numbers = []
for i in range(n):
    while True:
        try:
            num = float(input(f"Enter number {i+1}: "))
            numbers.append(num)
            break
        except ValueError:
            print("Invalid input! Please enter a numeric value.")

# Perform the selected operation
if choice == 1:
    result = calculate_sum(numbers)
    operation_name = "Sum"
elif choice == 2:
    result = calculate_average(numbers)
    operation_name = "Average"
elif choice == 3:
    result = calculate_max(numbers)
    operation_name = "Maximum"
else:
    result = calculate_min(numbers)
    operation_name = "Minimum"

print(f"{operation_name} of the numbers: {result}")
```

4.2 Exercise on List Manipulation:

```python
sample_list = [1, 2, 3, 4, 5, 6]

# 1. Extract every other element
def extract_every_other(lst):
    """Return every other element from the list, starting from the first."""
    return lst[::2]

# 2. Slice a Sublist:
def get_sublist(lst, start, end):
    """Return a sublist from start to end (inclusive)."""
    return lst[start:end+1]

# 3. Reverse a List Using Slicing:
def reverse_list(lst):
    """Return the reversed list using slicing."""
    return lst[::-1]

# 4. Remove the First and Last Elements:
def remove_first_last(lst):
    """Return the list without the first and last elements."""
    return lst[1:-1]

# 5. Get the First n Elements:
def get_first_n(lst, n):
    """Return the first n elements of the list."""
    return lst[:n]

# 6. Extract Elements from the End:
def get_last_n(lst, n):
    """Return the last n elements of the list."""
    return lst[-n:]

# 7. Extract Elements in Reverse Order:
def reverse_skip(lst):
    """Return every second element starting from the second-to-last, moving backward."""
    return lst[-2::-2]

print("Original list:", sample_list)
print("1. Every other element:", extract_every_other(sample_list))
print("2. Sublist from index 2 to 4:", get_sublist(sample_list, 2, 4))
print("3. Reversed list:", reverse_list(sample_list))
print("4. Remove first and last elements:", remove_first_last(sample_list))
print("5. First 3 elements:", get_first_n(sample_list, 3))
print("6. Last 2 elements:", get_last_n(sample_list, 2))
print("7. Reverse skip elements:", reverse_skip(sample_list))
```

4.3 Exercise on Nested List:

```python
nested_list = [[1, 2], [3, [4, 5]], [6, 2]]

# 1. Flatten a Nested List (one level):
def flatten(lst):
    """
    Flatten a nested list (one level).
    Here I used isinstance() to check if an element is a list.
    isinstance(item, list) : True if item is a list (or subclass of list)
    """
    result = []
    for item in lst:
        if isinstance(item, list):  # check if item is a list
            result.extend(item)
    return result

# 2. Accessing Nested List Elements:
def access_nested_element(lst, indices):
    """Access an element from a nested list using indices."""
    element = lst
    for idx in indices:
        element = element[idx]
    return element

# 3. Sum of All Elements in a Nested List:
def sum_nested(lst):
    """Return the sum of all numbers in a nested list (any depth).
    Here I used type() to check the type of item (simpler basic Python check).
    type(item) == list : True only if item is exactly a list
    """
    total = 0
    for item in lst:
        if type(item) == list:  # check if item is a list
            total += sum_nested(item)
        else:
            total += item
    return total

# 4. Remove Specific Element from a Nested List:
def remove_element(lst, elem):
    """Remove all occurrences of elem from a nested list (one level)."""
    result = []
    for item in lst:
        if type(item) == list:
            filtered = [x for x in item if x != elem]
            result.append(filtered)
        else:
            if item != elem:
                result.append(item)
    return result

# 5. Find the Maximum Element in a Nested List:
def find_max(lst):
    """Return the maximum element in a nested list (any depth)."""
    max_val = float('-inf')
    for item in lst:
        if type(item) == list:
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val

# 6. Count Occurrences of an Element in a Nested List:
def count_occurrences(lst, elem):
    """Count how many times elem appears in a nested list (any depth)."""
    count = 0
    for item in lst:
        if type(item) == list:
            count += count_occurrences(item, elem)
        else:
            if item == elem:  # using '==' to check equality
                count += 1
    return count

# 7. Flatten a List of Lists of Lists:
def deep_flatten(lst):
    """Flatten a deeply nested list into a single list (any depth)."""
    result = []
    for item in lst:
        if type(item) == list:
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

# 8. Nested List Average:
def average_nested(lst):
    """Return the average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print("Nested list:", nested_list)
print("1. Flatten:", flatten(nested_list))
print("2. Access element [1,1,0]:", access_nested_element(nested_list, [1,1,0]))
print("3. Sum nested:", sum_nested(nested_list))
print("4. Remove element 2:", remove_element(nested_list, 2))
print("5. Find max:", find_max(nested_list))
print("6. Count occurrences of 2:", count_occurrences(nested_list, 2))
print("7. Deep flatten:", deep_flatten(nested_list))
print("8. Nested average:", average_nested(nested_list))
```

```python
# Problem - 4: Matrix Operations:

# Matrix Proofs
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)

print("A inverse:\n", A_inv)
print("\nA * A^-1 =\n", identity)
print("Hence A.A^-1 = I proved\n")

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB =\n", AB)
print("\nBA =\n", BA)

if not np.array_equal(AB, BA):
    print("Hence AB ≠ BA proved\n")

# 3. Prove (AB)^T = B^T A^T
left = (AB).T
right = np.dot(B.T, A.T)

print("(AB)^T =\n", left)
print("\nB^T A^T =\n", right)

if np.array_equal(left, right):
    print("Hence (AB)^T = B^T A^T proved")
```

```python
# Solve Linear System Using Matrix Form

# Matrix A
A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

# Matrix B
B = np.array([-1, -3, 9])

# Solve AX = B
X = np.linalg.solve(A, B)

print("Solution using np.linalg.solve:")
print("[x, y, z] =", X)
```

```python
# Solve using np.linalg.inv

A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

B = np.array([-1, -3, 9])

# Inverse of A
A_inv = np.linalg.inv(A)

# X = A⁻¹B
X = np.dot(A_inv, B)

print("Solution using inverse method:")
print("[x, y, z] =", X)
```

Method 1 (np.linalg.solve()) : Directly solves AX = B


Method 2 (np.linalg.inv()) : Uses formula Uses formula X = A^-1 B

10.2 Experiment: How Fast is Numpy?

```python
import time
# 1. Element-wise Addition:
size = 1_000_000

# Python lists version
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
end = time.time()
print("Python list addition time:", end - start, "seconds")

# NumPy version
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
result_np = arr1 + arr2
end = time.time()
print("NumPy array addition time:", end - start, "seconds")
```

```python
# 2. Element-wise Multiplication

# Python lists version
start = time.time()
result_list = [list1[i] * list2[i] for i in range(size)]
end = time.time()
print("Python list multiplication time:", end - start, "seconds")

# NumPy version
start = time.time()
result_np = arr1 * arr2
end = time.time()
print("NumPy array multiplication time:", end - start, "seconds")
```

```python
# 3. Dot Product

# Python lists version
start = time.time()
dot_list = sum([list1[i] * list2[i] for i in range(size)])
end = time.time()
print("Python list dot product time:", end - start, "seconds")

# NumPy version
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()
print("NumPy dot product time:", end - start, "seconds")
```

```python
# 4. Matrix Multiplication
# Python lists version
matrix_size = 500
mat1 = [[i+j for j in range(matrix_size)] for i in range(matrix_size)]
mat2 = [[i-j for j in range(matrix_size)] for i in range(matrix_size)]

start = time.time()
result = [[sum(mat1[i][k]*mat2[k][j] for k in range(matrix_size))
           for j in range(matrix_size)] for i in range(matrix_size)]
end = time.time()
print("Python list matrix multiplication time:", end - start, "seconds")

# NumPy version
np_mat1 = np.array(mat1)
np_mat2 = np.array(mat2)

start = time.time()
np_result = np.dot(np_mat1, np_mat2)
end = time.time()
print("NumPy array matrix multiplication time:", end - start, "seconds")
```

From the experiment, we can see that NumPy operations are significantly faster than plain Python lists for large arrays and matrix computations. For example, element-wise addition and dot product were almost 40–50 times faster, and matrix multiplication was over 100 times faster with NumPy.

NumPy is faster because it does calculations on whole arrays at once, is written in fast C code, and stores data efficiently in memory.
