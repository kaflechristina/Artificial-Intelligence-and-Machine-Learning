Remove emojis

```python
def remove_emoji(string):
  """
  This function will replace the emoji in string with whitespace
  """
  emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r' ', string)
```

Remove unwanted characters

```python
def remove_unwanted_characters(document):
    """
    Removes unnecessary characters and formatting from text data.

    This includes user mentions, hashtags, punctuation,
    emojis, and extra whitespace.

    Parameters:
        document (str): Input text to be cleaned.

    Returns:
        str: Cleaned text document.
    """

    # Remove user mentions (e.g., @username)
    document = re.sub(r"@[A-Za-z0-9_]+", " ", document)

    # Remove hashtags
    document = re.sub(r"#[A-Za-z0-9_]+", "", document)

    # Remove punctuation and special characters
    document = re.sub(r"[^0-9A-Za-z ]", "", document)

    # Remove emojis
    document = remove_emoji(document)

    # Replace multiple spaces with a single space
    document = re.sub(r"\s+", " ", document)

    return document.strip()
```

### Remove Punctutations:

```python
from nltk.tokenize import RegexpTokenizer

def remove_punct(text):
    """
    Removes punctuation from text using regex-based tokenization.

    Parameters:
        text (str): Input raw text.

    Returns:
        list: Tokenized text containing only alphanumeric words.
    """

    # Tokenizer that keeps only words (removes punctuation)
    tokenizer = RegexpTokenizer(r"\w+")

    # Tokenize directly from string input
    tokens = tokenizer.tokenize(text)

    return tokens
```

### Remove StopWord:

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
custom_stopwords = ['@', 'RT']
stop_words.update(custom_stopwords)
```

```python
def remove_stopwords(text_tokens):
    """
    Removes stopwords from a list of tokenized words.

    Parameters:
        text_tokens (list): List of word tokens.

    Returns:
        list: Tokens with stopwords removed.
    """

    result_tokens = []

    # Filter out stopwords
    for token in text_tokens:
        if token not in stop_words:
            result_tokens.append(token)

    return result_tokens
```

## Text Normalization:

### Lemmatization:


```python
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')


def lemmatization(token_text):
    """
    Performs lemmatization on a list of tokens.

    Parameters:
        token_text (list): List of word tokens.

    Returns:
        list: Lemmatized tokens.
    """

    lemmatizer = WordNetLemmatizer()

    # Lemmatize each token (defaulting to verb form)
    lemmatized_tokens = [
        lemmatizer.lemmatize(token, pos='v') for token in token_text
    ]

    return lemmatized_tokens
```

### Stemming:


```python
from nltk.stem import PorterStemmer

def stemming(token_text):
    """
    Applies Porter Stemming to a list of tokens.

    Parameters:
        token_text (list): List of word tokens.

    Returns:
        list: Stemmed tokens.
    """

    porter = PorterStemmer()
    stemmed_tokens = []

    for word in token_text:
        stemmed_tokens.append(porter.stem(word))

    return stemmed_tokens
```

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

# Download NLTK resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer


def lower_order(text):
    """Convert text to lowercase."""
    return text.lower()


def remove_urls(text):
    """Remove URLs from text."""
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub('', text)


def remove_emoji(text):
    """Remove emojis from text."""
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(' ', text)


def remove_unwanted_characters(document):
    """
    Removes mentions, hashtags, punctuation, emojis, and extra whitespace.
    """
    document = re.sub(r"@[A-Za-z0-9_]+", " ", document)
    document = re.sub(r"#[A-Za-z0-9_]+", "", document)
    document = re.sub(r"[^0-9A-Za-z ]", "", document)

    document = remove_emoji(document)

    # Normalize whitespace (fixes multiple spaces properly)
    document = re.sub(r"\s+", " ", document)

    return document.strip()


def stemming(text):
    """Apply Porter Stemming to tokenized text."""
    porter = PorterStemmer()
    return [porter.stem(word) for word in text]


def lemmatization(text):
    """Apply WordNet Lemmatization to tokenized text."""
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in text]
```

```python
def text_cleaning_pipeline(dataset, rule="lemmatize"):
    """
    Preprocess text data through a series of cleaning steps.

    Args:
        dataset (str): Input text to be cleaned.
        rule (str): Either 'lemmatize' or 'stem' to choose normalization method.

    Returns:
        str: Cleaned and processed text.
    """
    # Handle non-string or NaN input
    if not isinstance(dataset, str):
        return ""

    # Convert to lowercase
    data = lower_order(dataset)

    # Remove URLs
    data = remove_urls(data)

    # Remove emojis
    data = remove_emoji(data)

    # Remove other unwanted characters (mentions, hashtags, punctuation)
    data = remove_unwanted_characters(data)

    # Tokenize
    tokens = word_tokenize(data)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Apply lemmatization or stemming
    if rule == "lemmatize":
        tokens = lemmatization(tokens)
    elif rule == "stem":
        tokens = stemming(tokens)
    else:
        print("Pick between lemmatize or stem")
        return ""

    # Join tokens back into a string
    return " ".join(tokens)
```
